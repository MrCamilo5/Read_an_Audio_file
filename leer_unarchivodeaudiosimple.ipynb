```
import wave 
import struct
import matplotlib.pyplot as plt
import numpy as np
sonido = wave.open('punch.wav','rb')

pak=sonido.readframes(sonido.getnframes())
unp=list(struct.iter_unpack('b',pak))
aud=[unp[i][0] for i in range(len(unp))]
aud_int=np.array(aud,dtype='b')+127
plt.plot(aud_int)

sonido.close()
```
